```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("births_day_month.csv")
```

```python
# Function to multiply the number of births by the number of days that were in that month (moths with 31 days) 
def cal_fun(i):
       return i*31
```

```python
data['january']=data['january'].apply(cal_fun)
data['march']=data['march'].apply(cal_fun)
data['may']=data['may'].apply(cal_fun)
data['july']=data['july'].apply(cal_fun)
data['august']=data['august'].apply(cal_fun)
data['october']=data['october'].apply(cal_fun)
data['december']=data['december'].apply(cal_fun)
```

```python
# Function to multiply the number of births by the number of days that were in that month (moths with 30 days) 
def cal_30(i):
       return i*30
```

```python
data['april']=data['april'].apply(cal_30)
data['june']=data['june'].apply(cal_30)
data['september']=data['september'].apply(cal_30)
data['november']=data['november'].apply(cal_30)
```

```python
# The days in February I do separately as there is a year where the number of days is 29 instead of 28
data['february']=data['february'].replace({493.71: 493.71*28})
data['february']=data['february'].replace({494.54: 494.54*28})
data['february']=data['february'].replace({479.75: 479.75*28})
data['february']=data['february'].replace({476.31: 476.31*29})
data['february']=data['february'].replace({487.68: 487.68*28})
```
